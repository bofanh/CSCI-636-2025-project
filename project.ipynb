```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import numpy as np

# === Step 1: Load and Clean Data ===

file_path = "yellow_tripdata_2025-01.parquet"

# Load the data
df = pd.read_parquet(file_path)

# Calculate total_trip_time and add to df
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
df['total_trip_time'] = (np.absolute(df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime'])).dt.total_seconds()


# print(df.columns)
df.head()

df = df.dropna(subset=['passenger_count'])
df.shape
```

```python
# Convert to datetime
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])

# Extract minutes since midnight
df['pickup_min_since_midnight'] = df['tpep_pickup_datetime'].dt.hour * 60 + df['tpep_pickup_datetime'].dt.minute
df['dropoff_min_since_midnight'] = df['tpep_dropoff_datetime'].dt.hour * 60 + df['tpep_dropoff_datetime'].dt.minute
```

```python
df.columns
```

```python
# Normalize to [0, 1] over 1440 minutes
df['pickup_time_norm'] = df['pickup_min_since_midnight'] / 1440
df['dropoff_time_norm'] = df['dropoff_min_since_midnight'] / 1440

# Apply sine and cosine
df['pickup_time_sin'] = np.sin(2 * np.pi * df['pickup_time_norm'])
df['pickup_time_cos'] = np.cos(2 * np.pi * df['pickup_time_norm'])
df['dropoff_time_sin'] = np.sin(2 * np.pi * df['dropoff_time_norm'])
df['dropoff_time_cos'] = np.cos(2 * np.pi * df['dropoff_time_norm'])
```

## give ground truth in  Sustainability Score and Class

```python
# Create fare_per_km
df['fare_per_km'] = df['fare_amount'] / (df['trip_distance'] + 1e-5)

# Compute sustainability score
df['sustainability_score'] = (
    1.5 * df['passenger_count'] -
    1.0 * df['fare_per_km'] -
    0.5 * df['total_trip_time'] / 600
)

# Bin the score into 3 classes
df['sustainability_class'] = pd.qcut(df['sustainability_score'], q=3, labels=[0,1,2]).astype(int)
```

```python
# prompt: visulize df['sustainability_class'] = pd.qcut(df['sustainability_score'], q=3, labels=[0,1,2]).astype(int)

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.countplot(x='sustainability_class', data=df)
plt.title('Distribution of Sustainability Classes')
plt.xlabel('Sustainability Class')
plt.ylabel('Count')
plt.show()
```

```python
from sklearn.utils import resample

# Separate by class
df0 = df[df['sustainability_class'] == 0]
df1 = df[df['sustainability_class'] == 1]
df2 = df[df['sustainability_class'] == 2]

# Find the minimum class count
min_count = min(len(df0), len(df1), len(df2))

# Downsample each to same count
df0_bal = resample(df0, replace=False, n_samples=min_count, random_state=42)
df1_bal = resample(df1, replace=False, n_samples=min_count, random_state=42)
df2_bal = resample(df2, replace=False, n_samples=min_count, random_state=42)

# Combine
df_balanced = pd.concat([df0_bal, df1_bal, df2_bal]).sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
# Sample only 1% from each class
df_balanced = (
    df_balanced.groupby('sustainability_class', group_keys=False)
          .apply(lambda x: x.sample(frac=0.01, random_state=42), include_groups=True)
          .reset_index(drop=True)
)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline

# Feature & target selection
X = df_balanced[[
    'PULocationID', 'DOLocationID',
    'passenger_count', 'trip_distance', 'fare_amount', 'total_trip_time',
    'pickup_time_sin', 'pickup_time_cos', 'dropoff_time_sin', 'dropoff_time_cos'
]]
y = df_balanced['sustainability_class']

# Categorical columns to One-Hot Encode
categorical_cols = ['PULocationID', 'DOLocationID']

# Numerical columns to Standard Scale
numeric_cols = ['passenger_count', 'trip_distance', 'fare_amount', 'total_trip_time',
                'pickup_time_sin', 'pickup_time_cos', 'dropoff_time_sin', 'dropoff_time_cos']
```

```python
# Preprocessing pipeline
preprocessor = ColumnTransformer(transformers=[
    ('num', StandardScaler(), numeric_cols),
    ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), categorical_cols)
])
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.utils import resample
from sklearn.metrics import classification_report, accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

```

```python

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42)

```

```python
print(len(X_train), len(y_train), len(X_test), len(y_test))
```

```python
# Classifiers to compare
models = {
    'LogisticRegression': LogisticRegression(max_iter=1000),
    'RandomForest': RandomForestClassifier(),
    'GradientBoosting': GradientBoostingClassifier(),
    'KNN': KNeighborsClassifier(),
    'MLP': MLPClassifier(max_iter=1000)
}
# Train & evaluate
results = {}

for name, model in models.items():
    print(f"\n🚀 Training model: {name}")

    clf = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', model)
    ])

    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)

    # Basic metrics
    acc = accuracy_score(y_test, y_pred)
    report_dict = classification_report(y_test, y_pred, output_dict=True)
    results[name] = {'accuracy': acc, 'f1_macro': report_dict['macro avg']['f1-score']}

    # Show metrics summary
    print(f"✅ {name} Accuracy: {acc:.4f}")
    print("📋 Classification Report:")
    print(classification_report(y_test, y_pred, digits=4))

    # Plot confusion matrix
    cm = confusion_matrix(y_test, y_pred, labels=[0, 1, 2])
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Low (0)', 'Medium (1)', 'High (2)'])
    disp.plot(cmap='Blues')
    plt.title(f'Confusion Matrix: {name}')
    plt.grid(False)
    plt.show()
```

```python
# Show results
print("\n📊 Final Results Summary:")
results_df = pd.DataFrame(results).T
results_df.sort_values("f1_macro", ascending=False, inplace=True)
print(results_df)

# Plot
results_df.plot(kind='barh', figsize=(10, 5))
plt.title("Model Comparison: Accuracy and F1-Macro (Sustainability Classification)")
plt.xlabel("Score")
plt.grid(True)
plt.tight_layout()
plt.show()
```
